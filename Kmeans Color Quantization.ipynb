```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
image_array = mpimg.imread("C:/Users/ovied/OneDrive/Desktop/luna2.jpg") 
```

```python
plt.figure(dpi = 150)
plt.imshow(image_array)
```

```python
image_array.shape
```

```python
(h,w,c) = image_array.shape #assigning variables the the shape of the array
```

```python
h
```

```python
image_array2D = image_array.reshape(h*w,c)  #making the array 2D
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters= 10) #Quantize to 10 averaged out colors
label = model.fit_predict(image_array2D)
```

```python
rgb = model.cluster_centers_.round(0).astype(int)    #rounding the color codes and converting them to int.
rgb
```

```python
quantized_image = np.reshape(rgb[label], (h,w,c))
```

```python
plt.figure(figsize=(6,6),dpi=200)
plt.imshow(quantized_image)
```

```python
def imageQuan(path, kClusters):
    '''This function turns the image into a 3D array then converts it into a 2D array. Then it forms the desired amount of
    clusters before reconfiguring the 2D array into a 3D array. Lastly it plots the original image and the Quantized image '''
    image_array = mpimg.imread(path) 
    (h,w,c) = image_array.shape  #assigns variables to the height, width, and color
    image_array2D = image_array.reshape(h*w,c)  
    model = KMeans(n_clusters= kClusters) #Quantize to the amount of Colors selected by the clusters. 2 clusters = 2 Colors
    label = model.fit_predict(image_array2D)
    rgb = model.cluster_centers_.round(0).astype(int)
    quantized_image = np.reshape(rgb[label], (h,w,c))
    plt.figure(figsize=(6,6),dpi=200)
    plt.imshow(image_array)
    plt.figure(figsize=(6,6),dpi=200)
    plt.imshow(quantized_image)


```

```python
imageQuan("C:/Users/ovied/OneDrive/Desktop/sunset.jpg", 4 )
```
